```python
# Dynamic programming

# Optimization by caching

# Memoization / Caching

# Remember the solution to a problem
```

```python
def add_to_80(n):
    return n + 80

add_to_80(5)
```

```python
def memoized_add_to_80():
    # if the parameter doesn't changes
    # is memoized
    cache = {}
    def memoized(n):
        if (n in cache): 
            return cache[n]
        print("long time")
        cache[n] = n + 80
        return cache[n]

    return memoized

memoized = memoized_add_to_80()

print(memoized(5))
print(memoized(10))
print(memoized(5))
print(memoized(10))
```
